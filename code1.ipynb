```python
import pandas as pd
import numpy as np
import re
import nltk
import spacy
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics.pairwise import cosine_similarity

# Load NLP Model
nlp = spacy.load("en_core_web_sm")

# Load Dataset
df = final_df # Change filename accordingly

# ========== 1️⃣ CLEAN TEXT COLUMNS ==========
def clean_text(text):
    """Removes special characters, numbers, and stopwords"""
    if pd.isnull(text): return ""
    text = re.sub(r'\W+', ' ', text)  # Remove special characters
    text = text.lower().strip()  # Convert to lowercase
    return text

df["resume_text"] = df["resume_text"].apply(clean_text)
df["job_description"] = df["job_description"].apply(clean_text)
df["required_skills"] = df["required_skills"].apply(clean_text)

# ========== 2️⃣ EXTRACT SKILLS FROM RESUMES ==========
def extract_skills(text):
    """Extracts skills from resume using NLP"""
    skills = []
    doc = nlp(text)
    for token in doc:
        if token.pos_ in ["NOUN", "PROPN"]:  # Common for skill names
            skills.append(token.text)
    return list(set(skills))  # Remove duplicates

df["resume_skills"] = df["resume_text"].apply(extract_skills)

# ========== 3️⃣ CLEAN EXPERIENCE COLUMN ==========
def extract_experience(exp):
    """Converts '2 to 9 Years' → min=2, max=9"""
    match = re.findall(r'\d+', str(exp))
    if len(match) == 2:
        return int(match[0]), int(match[1])
    elif len(match) == 1:
        return int(match[0]), int(match[0])  # Same min-max if only one number
    return np.nan, np.nan  # Handle missing values

df[['min_experience', 'max_experience']] = df['experience'].apply(lambda x: pd.Series(extract_experience(x)))

# ========== 4️⃣ CONVERT TEXT TO NUMERIC FEATURES ==========
# 🔹 Encode Job Titles
le = LabelEncoder()
df["encoded_job_title"] = le.fit_transform(df["job_title"])

# 🔹 Convert Job Descriptions to TF-IDF Vectors
vectorizer = TfidfVectorizer(stop_words="english", max_features=500)
job_desc_matrix = vectorizer.fit_transform(df["job_description"])

# 🔹 Convert Resume Text to TF-IDF Vectors
resume_matrix = vectorizer.transform(df["resume_text"])

# 🔹 Compute Cosine Similarity (Resume ↔ Job Description)
df["match_score"] = [cosine_similarity(resume_matrix[i], job_desc_matrix[i])[0][0] for i in range(len(df))]

# 🔹 Final Processed Data
df_processed = df[["resume_text", "resume_skills", "job_title", "encoded_job_title", 
                    "min_experience", "max_experience", "match_score"]]

# Save Processed Data
df_processed.to_csv("processed_dataset.csv", index=False)

print("✅ Data Preprocessing Complete!")
```

```python
import pandas as pd
import numpy as np
import re
import nltk
import spacy
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics.pairwise import cosine_similarity
from gensim.models import Word2Vec
from sentence_transformers import SentenceTransformer

nltk.download('stopwords')
from nltk.corpus import stopwords

# Load datasets
final_df = pd.read_csv(r"C:\Users\Admin\Downloads\clean_combined_dataset.csv")

# 🔹 1. **Text Preprocessing**
def preprocess_text(text):
    if pd.isnull(text):
        return ""
    text = re.sub(r'<.*?>', '', text)  # Remove HTML tags
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    text = re.sub(r'[^\w\s]', '', text)  # Remove special characters
    text = text.lower()  # Convert to lowercase
    stop_words = set(stopwords.words('english'))
    text = " ".join(word for word in text.split() if word not in stop_words)  # Remove stopwords
    return text

final_df['resume_text'] = final_df['resume_text'].apply(preprocess_text)
final_df['job_description'] = final_df['job_description'].apply(preprocess_text)
final_df['required_skills'] = final_df['required_skills'].astype(str).apply(preprocess_text)

# 🔹 2. **TF-IDF Vectorization**
tfidf = TfidfVectorizer(max_features=5000)
resume_tfidf = tfidf.fit_transform(final_df['resume_text'])
job_tfidf = tfidf.transform(final_df['job_description'])

# 🔹 3. **Cosine Similarity (Resume vs. Job Description)**
final_df['similarity_score'] = [cosine_similarity(resume_tfidf[i], job_tfidf[i])[0][0] for i in range(len(final_df))]

# 🔹 4. **Word2Vec Embeddings**
sentences = [text.split() for text in final_df['resume_text']]
w2v_model = Word2Vec(sentences, vector_size=100, window=5, min_count=2, workers=4)
def get_w2v_vector(text):
    words = text.split()
    vector = np.mean([w2v_model.wv[word] for word in words if word in w2v_model.wv] or [np.zeros(100)], axis=0)
    return vector

final_df['resume_w2v'] = final_df['resume_text'].apply(get_w2v_vector)

# 🔹 5. **BERT Embeddings (Sentence Transformers)**
bert_model = SentenceTransformer('all-MiniLM-L6-v2')
final_df['resume_bert'] = final_df['resume_text'].apply(lambda x: bert_model.encode(x))
final_df['job_bert'] = final_df['job_description'].apply(lambda x: bert_model.encode(x))

# Save Processed Data
final_df.to_csv(r"C:\Users\Admin\Downloads\feature_engineered_dataset.csv", index=False)
print("✅ Feature Engineering Completed and Saved!")
```

```python
import os
import pandas as pd
import numpy as np
import streamlit as st
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
import fitz  # PyMuPDF

# Load sentence transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Function to extract text from PDF
def extract_text_from_pdf(file):
    with fitz.open(stream=file.read(), filetype="pdf") as doc:
        text = ""
        for page in doc:
            text += page.get_text()
    return text

# Sidebar
st.sidebar.title("NLP Quick CV Selector")
num_top = st.sidebar.slider("Select Top K Candidates", 1, 20, 5)

# Upload job description
st.subheader("Upload Job Description")
job_desc_file = st.file_uploader("Upload Job Description PDF", type=["pdf"], key="job_desc")

# Upload multiple CVs
st.subheader("Upload Candidate CVs")
cv_files = st.file_uploader("Upload CV PDFs", type=["pdf"], accept_multiple_files=True, key="cvs")

if job_desc_file and cv_files:
    with st.spinner("Processing..."):

        # Extract job description text
        job_desc_text = extract_text_from_pdf(job_desc_file)
        job_embedding = model.encode([job_desc_text])[0]

        # Extract and embed CVs
        cv_data = []
        for cv_file in cv_files:
            name = cv_file.name
            text = extract_text_from_pdf(cv_file)
            embedding = model.encode([text])[0]
            similarity = cosine_similarity([job_embedding], [embedding])[0][0]
            cv_data.append((name, similarity, text))

        # Sort CVs by similarity
        sorted_cv_data = sorted(cv_data, key=lambda x: x[1], reverse=True)

        # Display top K candidates
        st.success("Top Matched CVs:")
        for i, (name, score, text) in enumerate(sorted_cv_data[:num_top], start=1):
            st.markdown(f"### {i}. {name}")
            st.markdown(f"**Match Score:** {score:.2f}")
            with st.expander("Show CV Text"):
                st.write(text[:2000] + "..." if len(text) > 2000 else text)

# Footer
st.sidebar.markdown("---")
st.sidebar.info("Built with ❤️ using Streamlit and SentenceTransformers")

```

```python
from sentence_transformers import SentenceTransformer
import torch

# Load lightweight and fast model
bert_model = SentenceTransformer('paraphrase-MiniLM-L3-v2')

# Use GPU if available
device = 'cuda' if torch.cuda.is_available() else 'cpu'
bert_model = bert_model.to(device)

# Encode with progress bar and batching
resume_embeddings = bert_model.encode(
    final_df['resume_text'].tolist(),
    batch_size=32,                      # Reasonable speed vs memory
    show_progress_bar=True,
    device=device
)

job_desc_embeddings = bert_model.encode(
    final_df['job_description'].tolist(),
    batch_size=32,
    show_progress_bar=True,
    device=device
)
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Function to compute similarity in batches
def compute_diagonal_similarity(resume_embeddings, job_desc_embeddings, batch_size=1000):
    num_samples = len(resume_embeddings)
    similarity_scores = np.zeros(num_samples)  # Store only diagonal values

    for i in range(0, num_samples, batch_size):
        end = min(i + batch_size, num_samples)
        batch_similarity = cosine_similarity(resume_embeddings[i:end], job_desc_embeddings[i:end])
        similarity_scores[i:end] = np.diag(batch_similarity)  # Only store diagonal
        
    return similarity_scores

# Compute similarity efficiently
final_df['resume_job_similarity'] = compute_diagonal_similarity(resume_embeddings, job_desc_embeddings)

# Preview results
print(final_df[['resume_text', 'job_description', 'resume_job_similarity']].head())
```

```python
import streamlit as st
import pandas as pd
import torch
import torch.nn.functional as F
from transformers import BertTokenizer, BertForSequenceClassification
import base64

# Load BERT model
@st.cache_resource
def load_model():
    tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
    model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)
    model.eval()
    return tokenizer, model

tokenizer, model = load_model()

# Encode function
def bert_encode(text):
    return tokenizer(text, padding=True, truncation=True, return_tensors="pt", max_length=512)

# Predict similarity score
def predict_match_score(resume, job_desc):
    inputs = bert_encode(resume + " [SEP] " + job_desc)
    with torch.no_grad():
        outputs = model(**inputs)
        scores = F.softmax(outputs.logits, dim=1)
        return float(scores[0, 1])  # match probability

# Sidebar
st.sidebar.title("📁 Upload Files")
uploaded_resumes = st.sidebar.file_uploader("Upload multiple resumes (TXT only)", type=['txt'], accept_multiple_files=True)
uploaded_jd = st.sidebar.file_uploader("Upload Job Description", type=['txt'])

st.title("🔍 Resume Matcher & Job Recommender")

# If both resumes and JD are uploaded
if uploaded_resumes and uploaded_jd:
    jd_text = uploaded_jd.read().decode('utf-8')

    resume_data = []
    for file in uploaded_resumes:
        resume_text = file.read().decode('utf-8')
        score = predict_match_score(resume_text, jd_text)
        resume_data.append({
            "Resume Name": file.name,
            "Match Score": round(score, 4),
            "Content": resume_text
        })

    # DataFrame with scores
    df = pd.DataFrame(resume_data)
    df = df.sort_values("Match Score", ascending=False)

    st.success("✅ Resumes ranked by job description match!")
    st.dataframe(df[["Resume Name", "Match Score"]].reset_index(drop=True))

    # Top K selector
    top_k = st.slider("🎯 Select top K candidates", min_value=1, max_value=len(df), value=3)
    st.subheader(f"📄 Top {top_k} Resumes Preview")
    for i, row in df.head(top_k).iterrows():
        with st.expander(f"{row['Resume Name']} - Score: {row['Match Score']}"):
            st.text(row['Content'][:1500])  # show first 1500 characters

    # Job Role Recommendation (basic logic)
    st.subheader("💡 Suggested Job Role Based on Top Resume")
    if not df.empty:
        top_resume = df.iloc[0]["Content"].lower()
        if "data scientist" in top_resume:
            st.info("Recommended Role: Data Scientist")
        elif "python" in top_resume and "ml" in top_resume:
            st.info("Recommended Role: ML Engineer")
        elif "sql" in top_resume and "dashboard" in top_resume:
            st.info("Recommended Role: Data Analyst")
        else:
            st.info("Recommended Role: Software Developer")

else:
    st.warning("👈 Please upload resumes and a job description to begin.")
```
